```python
df.isnull().sum()
#with what to fill the missing values? MEAN?? MEDIAN??
#what we choose is median,but there is a catch-overall median nahi lenge-we take median of each category
#and then put it in
#df["Review Rating"][1]
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(0))
#It fills missing values (NaN) in Review Rating using the median rating of the same Category.
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels=['Young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days

frequency_mapping={
    'Fortnightly':14,'Weekly':7,'Monthly':30,'Quarterly':90,'Bi-Weekly':14,'Annually':365,'Every 3 months':90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
